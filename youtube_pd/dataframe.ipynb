```python
import numpy as np
import pandas as pd
```

```python
# create DataFrame (with random values)

newdf = pd.DataFrame(np.random.rand(340,5))
```

```python
# read

newdf
```

```python
# read limited

newdf.head()
```

```python
# type

type(newdf)
```

```python
# basic stats

newdf.describe()
```

```python
# check data types
# if one value's type is differnet then dtype = object

newdf.dtypes
```

```python
# print indices count

newdf.index
```

```python
# columns count

newdf.columns
```

```python
# create numpy of dataframe

newdf.to_numpy()
```

```python
# transpose : rows as columns & columns as rows

newdf.T
```

```python
# sort index
# axis = 0 : rows
# axis = 1 : columns
# by default (ascending = True)
# ascending = Flase : reverse sorting

newdf.sort_index(axis=0,ascending=False)
```

```python
# one value

newdf.at[3,3] 
```

```python
# one column

newdf[3]
```

```python
# view : pointer
# newdf1 POINTS to same memory loc as newdf
# changing newdf1 changes newdf

newdf1 = newdf
```

```python
newdf1.at[1,0] = 98
```

```python
# newdf1.loc[1,0] = 98 same
```

```python
# copy : doesnt affect original jst creates copy
# newdf2 = newdf[:] : same

newdf2 = newdf.copy()
```

```python
newdf2.at[3,3] = 333333
```

```python
newdf
```

```python
# change columns name 

newdf.columns = list("ABCDE")
```

```python
# add new column

newdf.loc[0,'F'] = 9812
```

```python
# delete one column

newdf = newdf.drop('F',axis = 1)
```

```python
# specific rows & columns

newdf.loc[[3,4,5],['A','E']]
```

```python
# condition access

newdf[(newdf['B'] < 0.2)]
```

```python
# multiple conditions

newdf[(newdf['A'] < 0.2) & (newdf['D'] > 0.8)]
```

```python
# integer based access
# no matter rows & columns it takes index

newdf.iloc[2]
```

```python
# accessing multiple rows

newdf.iloc[34:48]
```

```python
# drop row / rows

newdf.drop(5)

# newdf.drop([1,2,3])
```

```python
# drop column/columns

newdf.drop(['C','E'],axis = 1)
```

```python
# inplace : modifies the original df

newdf.drop([7,8],inplace=True)
```

```python
newdf.head(10)
```

```python
# reset index : gives both previous & new index 

newdf.reset_index()
```

```python
# drop old index : drops 'index' column

newdf.reset_index(drop=True)
```

```python
# look for 'null' values

newdf.isnull()
```

```python
# look for null in individual row

newdf.loc[10].isnull()
```

```python
# look for null in individual column

newdf['C'].isnull()
```

```python
# change all values to NONE (or any value) in column

newdf['B'] = None
```

```python
# change all values to NONE (or any value) in row

newdf.loc[6] = 65
```

```python
# change values in some index

newdf.loc[335:339,'E'] = 5.005
```

```python
# drop Na ( NaN / NaT ...)

newdf.dropna()
```

```python
# drop duplicates

newdf.drop_duplicates(subset = ['E'])
```

```python
# keep first (def) , last , False

newdf.drop_duplicates(subset = ['B'] , keep = 'last')
```

```python
newdf.drop_duplicates(subset = ['B'] , keep = False)
```

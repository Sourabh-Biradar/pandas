```python
# pandas 

import pandas as pd

weather_data = [
    ('2025-05-01', 'New York', 18, 'Sunny'),
    ('2025-05-01', 'Los Angeles', 22, 'Cloudy'),
    ('2025-05-01', 'Chicago', 15, 'Rainy'),
    ('2025-05-01', 'Houston', 28, 'Sunny'),
    ('2025-05-01', 'Phoenix', 33, 'Hot'),
    ('2025-05-02', 'Seattle', 16, 'Cloudy'),
    ('2025-05-02', 'Miami', 29, 'Sunny'),
    ('2025-05-02', 'Denver', 14, 'Snow'),
    ('2025-05-02', 'Boston', 17, 'Windy'),
    ('2025-05-02', 'San Francisco', 19, 'Foggy'),
    ('2025-05-03', 'Las Vegas', 32, 'Hot'),
    ('2025-05-03', 'Atlanta', 25, 'Rainy'),
    ('2025-05-03', 'Dallas', 27, 'Sunny'),
    ('2025-05-03', 'Philadelphia', 20, 'Cloudy'),
    ('2025-05-03', 'Orlando', 30, 'Sunny'),
    ('2025-05-04', 'Detroit', 13, 'Rainy'),
    ('2025-05-04', 'San Diego', 21, 'Sunny'),
    ('2025-05-04', 'Minneapolis', 12, 'Snow'),
    ('2025-05-04', 'Portland', 15, 'Cloudy'),
    ('2025-05-04', 'Austin', 26, 'Humid')
]

df = pd.DataFrame(weather_data,columns=['Date','City','Temperature','Event'])
df

# DataFrame is datatype in pandas , basically its Table
# syntax : pd.DataFrame(readDataFrom,columns=[])
# columns : lets you name columns
```

```python
print(df.shape)     # rows x columns

print(df.head())    # first 5 rows
print(df.tail())    # bottom 5 rows

print(df.head(7))   # first specified rows
print(df.tail(10))  # bottom specified rows

print(df[5:8])      # slicing

print(df.columns) # lists all column names
```

```python
print(df.City)          # single column

print(df['City'])       # same , single column

print(df[['Date','Event']])  # multiple columns
```

```python
print(df['Temperature'].max())   # min(), mean(), std(), count() can use these too

print(df['Temperature'].describe())  # lists all min(),min(), mean(), std(), count()...
```

```python
maxT = df[df.Temperature==df.Temperature.max()]    # df[df.Temperature==33]
print(maxT)

minDate = df[df.Date==df.Date.min()]               # all rows with minimum Date
print(minDate)

minT  = df[df.Temperature==df.Temperature.min()].City   # single value
print('City with minimum Temperature :',minT)
```
